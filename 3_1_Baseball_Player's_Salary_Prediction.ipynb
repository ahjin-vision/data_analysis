## 회귀분석에 사용할 feature 확인하기 

### 수치형자료: 단위 정규화

```
# 판다스 형태로 정의된 데이터 출력시 float모양으로 출력되게 
pd.options.mode.chained_assignment = None

def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean) / series_std)
    return df

scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR',
       '연봉(2018)', '연봉(2017)']
pitcher_df = standard_scaling(pitcher, scale_columns)
display(pitcher_df.head(10))
```

```
# 수치형 데이터만! 
pitcher_features = pitcher_df[['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR',
       '연봉(2018)', '연봉(2017)']]

# 각 feautre의 히스토그램
def plot_hist_each_columns(df):
    plt.rcParams['figure.figsize'] = [20, 16]
    fig = plt.figure(1)

    for i in range(len(df.columns)):
        ax = fig.add_subplot(4, 5, i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()

plot_hist_each_columns(pitcher_features)
```

### 범주형자료: 원핫인코딩

```
team_encoding = pd.get_dummies(pitcher_df['팀명'])
display(team_encoding.head(5))

pitcher_df = pitcher_df.drop('팀명', axis=1)
pitcher_df = pitcher_df.join(team_encoding)
display(pitcher_df.head(10))
```

# 회귀분석

## 회귀분석에 사용할 훈련데이터셋과 테스트데이터셋 준비하기

```
pitcher_df = pitcher_df.rename(columns={'연봉(2018)':'y'})
```

```
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

X = pitcher_df[pitcher_df.columns.difference(['선수명', 'y'])]
y = pitcher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## 학습

```
# 회귀분석계수
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

print(lr.coef_)
```

## 평가
1. 수정 결정 계수(R2 score)
2. RMSE
2. statsmodel의 summary( )
3. coef 시각화

### R2 score

```
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

훈련데이터셋 점수가 테스트데이터셋 점수보다 높으므로 과적합이 일어남. 

### RMSE
- Root Mean Square Error
- 실제 값과 예측값의 차이를 절대적인 수치로 나타냄. 
- 값이 높을수록 예측이 부정확함. 

```
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions)))

y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions)))
```

### statsmodel의 summary( )

```
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train).fit()
model.summary()
```

<statsmodel의 summary를 이용한 유의미한 feature>

WAR, 연봉(2017), 한화

### 다중공선성 확인
회귀분석은 feature간 '독립성'을 전제로 한 분석방법이기 때문에 올바른 분석을 위해서는 변수간 상관관계가 높은 feature 쌍을 제거해주는 것이 좋다. 

다중공선성이란 변수간 상관관계가 높아 분석에 부정적인 영향을 미치는 것을 의미한다. 

다중공선성은 분산팽창요인(Variance Inflation Factor, VIF)으로 평가할 수 있는데, 10~15를 넘으면 해당 feature은 다중공선성의 문제가 발생했다고 판단한다. 

```
from statsmodels.stats.outliers_influence import variance_inflation_factor

# feature마다 VIF 계수를 출력한다. 
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```

<VIF를 통해 강한 상호연관성을 가진 feature> --제거해줄것! 

FIP, kFIP

<coef 시각화를 통해 알아본 영향력이 강한 feature>

FIP, WAR, 홈런, 연봉(2017)

<statsmodel의 summary를 이용한 유의미한 feature>

WAR, 연봉(2017), 한화

**<<적절한 feature 선정하기>>**
1. VIF 계수가 높은 feature 제거: (FIP, kFIP와 같이 유사한 feature은 하나만 제거)
2. 다시 공선성 검증: 

    이전 단계에서는 VIF 계수가 높았던 볼넷, 삼진 등의 피처의 VIF 계수가 낮아진 것을 확인할 수 있음. 
VIF 계수가 비정상적으로 높은 feature을 제거하면 다른 feature들의 공선성도 자연스레 감소하기 때문.
3. 여전히 VIF 계수가 높은 feature 제거
4. 남은 feature를 토대로 다시한번 회귀분석 실시:  

    분석 결과에서 p-value가 유의하면서도 영향력이 큰 feature들을 선정.

=> 'FIP', 'WAR', '볼넷/9', '삼진/9', '연봉(2017)'

### 적절한 feature로 다시 학습하기


```
# 새로운 훈련데이터셋과 테스트 데이터셋 나누기 
X = pitcher_df[['FIP', 'WAR', '볼넷/9', '삼진/9', '연봉(2017)']]
y = pitcher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
